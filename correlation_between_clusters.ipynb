I'm going to take the clusters and show each of the graphs and the time series to show the correlation between the stocks


here is how the code should work

for cluster 0 to 14 I want to find stocks that belong in the current cluster
   for the current cluster, go through and find the csv file for that ticker that belong in cluster i
   use the close price to graph the movement of price onto the graph that corresponds to the current cluster i

do this for every single cluster 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define paths
clustered_data_path = 'dataset/stock_clusters.csv'
stock_folder = 'dataset/stock_csv/'

# Load clustered data
clustered_data = pd.read_csv(clustered_data_path)

# Define the stock column name and ensure it's correct
stock_column = 'Ticker'  # Change to the actual column name if different
cluster_column = 'Cluster'  # Ensure this matches the column name in clustered_data.csv

# Set the range for clusters (0 to 14, as specified)
num_clusters = 15

# Iterate through each cluster
for cluster in range(num_clusters):
    # Select stocks that belong to the current cluster
    cluster_stocks = clustered_data[clustered_data[cluster_column] == cluster]
    
    # Check if there are any stocks in this cluster
    if cluster_stocks.empty:
        print(f"No stocks found for Cluster {cluster}")
        continue
    
    # Create a new plot for the current cluster
    plt.figure(figsize=(12, 6))
    print(f"Processing Cluster {cluster} with stocks: {cluster_stocks[stock_column].tolist()}")
    
    # Loop through each stock in the current cluster
    for stock in cluster_stocks[stock_column]:
        stock_file = os.path.join(stock_folder, f"{stock}.csv")
        
        # Check if the stock's time series data file exists
        if not os.path.exists(stock_file):
            print(f"File for stock {stock} not found in {stock_folder}")
            continue
        
        try:
            # Load stock data and ensure the 'Close' column exists
            stock_data = pd.read_csv(stock_file)
            stock_data['Date'] = pd.to_datetime(stock_data['Date'])
            stock_data = stock_data.sort_values('Date')
            
            if 'Close' not in stock_data.columns:
                print(f"'Close' column missing in {stock_file}")
                continue
            
            # Plot the close price over time for this stock
            plt.plot(stock_data['Date'], stock_data['Close'], label=stock)
        
        except Exception as e:
            print(f"Error processing {stock_file}: {e}")
            continue
    
    # Customize and show plot for the current cluster
    plt.title(f'Time Series for Cluster {cluster}')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend(title="Stocks")
    plt.show()
```
