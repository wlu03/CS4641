1. **Date**: The trading date for the stock data.

2. **Open**: The price at which the stock started trading for the day.

3. **High**: The highest price the stock reached during the trading day.

4. **Low**: The lowest price the stock reached during the trading day.

5. **Close**: The final price of the stock at the end of the trading day.

6. **Volume**: The total number of shares traded during the day.

7. **OpenInt (Open Interest)**: Commonly used in options and futures, it represents the total number of outstanding contracts that have not been settled.

8. **Daily Return (%)**: The percentage change in the stock’s closing price from the previous day’s close, calculated as:
   \[
   \text{Daily Return} = \frac{\text{Close} - \text{Previous Close}}{\text{Previous Close}} \times 100
   \]

9. **Moving Average (MA-5, MA-10, MA-20)**: The average of the closing prices over a set period of days (5, 10, or 20). Moving averages help smooth out price data to identify trends.

10. **EMA (Exponential Moving Average)**: Similar to a simple moving average but gives more weight to recent prices, making it more responsive to new information.

11. **Bollinger Bands (BB_upper, BB_lower)**: Bands plotted above and below a moving average, typically two standard deviations away, used to assess price volatility. The upper band is the moving average + 2 standard deviations, and the lower band is the moving average - 2 standard deviations.

12. **Volatility-10**: A measure of the stock's price variation over a 10-day period, often calculated as the standard deviation of daily returns over this period.

13. **RSI (Relative Strength Index)**: A momentum oscillator that measures the speed and change of price movements, typically over a 14-day period, and ranges from 0 to 100. It’s used to identify overbought or oversold conditions.

14. **Volume Moving Average (Volume MA-20)**: The average volume traded over the past 20 days, helping identify changes in trading activity.

15. **High-Low (%)**: The percentage difference between the high and low prices of the day, calculated as:

    $$\frac{\text{High} - \text{Low}}{\text{Low}} \times 100$$

16. **Close-Open (%)**: The percentage difference between the closing and opening prices of the day, calculated as:
    
    $$\frac{\text{Close} - \text{Open}}{\text{Open}} \times 100$$

17. **Rate of Change over 10 days (ROC-10)**: The percentage change in price compared to the price 10 days ago, indicating momentum over this period.

18. **P/B Ratio (Price-to-Book Ratio)**: The ratio of a stock’s market price to its book value, indicating whether it’s undervalued or overvalued compared to its actual asset value.

19. **P/E Ratio (Price-to-Earnings Ratio)**: The ratio of a stock’s current price to its earnings per share (EPS), commonly used to assess valuation relative to earnings.

20. **Log Returns**: The logarithmic return of the stock price, calculated as:

    $$\ln\left(\frac{\text{Close}}{\text{Previous Close}}\right)$$
    Log returns are useful for analyzing continuous compounding over time.

21. **Dividend Yield**: The ratio of a company’s annual dividend payments to its share price, expressed as a percentage. It indicates the return on investment from dividends alone.

22. **Sector**: The industry sector to which the stock belongs (e.g., Technology, Healthcare), useful for sector-based analysis or comparisons.


```python
import os
import pandas as pd
import numpy as np
import yfinance as yf

# Path to your folder with CSV files
input_folder = 'dataset/stock_csv'

# Helper function to get financial metrics from Yahoo Finance
def get_financial_metrics(ticker):
    stock = yf.Ticker(ticker)
    info = stock.info
    
    # Retrieve specific financial metrics, with default None if not available
    pb_ratio = info.get('priceToBook', None)
    pe_ratio = info.get('trailingPE', None)
    dividend_yield = info.get('dividendYield', None)
    sector = info.get('sector', None)
    
    # Convert dividend yield to percentage if it's not None
    if dividend_yield is not None:
        dividend_yield = dividend_yield * 100  # Convert to percentage

    return pb_ratio, pe_ratio, dividend_yield, sector

# Process each CSV file in the folder
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(input_folder, filename)
        
        # Load the existing CSV file into a DataFrame
        stock_df = pd.read_csv(file_path)
        
        # Daily Return
        stock_df['Daily Return'] = stock_df['Close'].pct_change() * 100
        
        # Moving Averages
        stock_df['MA-5'] = stock_df['Close'].rolling(window=5).mean()
        stock_df['MA-10'] = stock_df['Close'].rolling(window=10).mean()
        stock_df['MA-20'] = stock_df['Close'].rolling(window=20).mean()
        
        # EMA
        stock_df['EMA'] = stock_df['Close'].ewm(span=20, adjust=False).mean()
        
        # Bollinger Bands
        rolling_mean = stock_df['Close'].rolling(window=20).mean()
        rolling_std = stock_df['Close'].rolling(window=20).std()
        stock_df['BB_upper'] = rolling_mean + (rolling_std * 2)
        stock_df['BB_lower'] = rolling_mean - (rolling_std * 2)
        
        # Volatility-10
        stock_df['Volatility-10'] = stock_df['Daily Return'].rolling(window=10).std()
        
        # RSI
        delta = stock_df['Close'].diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
        rs = gain / loss
        stock_df['RSI'] = 100 - (100 / (1 + rs))
        
        # Volume Moving Average
        stock_df['Volume MA-20'] = stock_df['Volume'].rolling(window=20).mean()
        
        # High-Low (%)
        stock_df['High-Low'] = ((stock_df['High'] - stock_df['Low']) / stock_df['Low']) * 100
        
        # Close-Open (%)
        stock_df['Close-Open'] = ((stock_df['Close'] - stock_df['Open']) / stock_df['Open']) * 100
        
        # Rate of Change over 10 days
        stock_df['ROC-10'] = stock_df['Close'].pct_change(periods=10) * 100
        
        # Log Returns
        stock_df['Log Returns'] = np.log(stock_df['Close'] / stock_df['Close'].shift(1))
        
        # Assume filename is in the format "TICKER.csv" and extract ticker symbol
        ticker_symbol = filename.split('.')[0]
        
        # Fetch financial metrics from Yahoo Finance for the given ticker symbol
        pb_ratio, pe_ratio, dividend_yield, sector = get_financial_metrics(ticker_symbol)
        
        # Add financial metrics to each row in the DataFrame
        stock_df['P/B Ratio'] = pb_ratio
        stock_df['P/E Ratio'] = pe_ratio
        stock_df['Dividend Yield'] = dividend_yield
        stock_df['Sector'] = sector
        
        # Save the updated DataFrame back to CSV, overwriting the original file
        stock_df.to_csv(file_path, index=False)
        
        print(f"Updated {filename} with calculated features and Yahoo Finance metrics.")

print("All files have been updated with new features and Yahoo Finance data!")
```
